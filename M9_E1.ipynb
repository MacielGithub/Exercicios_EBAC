# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
# Fazer um dataframe contendo média de idade por sexo

municipios = sinasc_raw['munResNome'].unique()
medias_idade_municipio = pd.DataFrame(columns=['Municipio', 'Media_idade_mae', 'Media_idade_pai'])

for municipio in municipios:

    # Separa um df contendo apenas dado do município do loop
    df_municipio = pd.DataFrame(sinasc_raw[sinasc_raw['munResNome'] == municipio])

    # Calcula médias de idade de homens e mulheres
    media_idade_mae_municipio = df_municipio['IDADEMAE'].mean()
    media_idade_pai_municipio = df_municipio['IDADEPAI'].mean()

    # Cria linha a ser adicionada. Essa linha corresponde às medias de idade no município
    nova_linha = {
        'Municipio': municipio,
        'Media_idade_mae': media_idade_mae_municipio,
        'Media_idade_pai': media_idade_pai_municipio
        }

    # adiciona à lista final
    medias_idade_municipio = pd.concat([medias_idade_municipio, pd.DataFrame([nova_linha])], ignore_index=True)

medias_idade_municipio
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
datas = sinasc_raw['DTNASC'].unique()

data_pesomedio_sexo_escolaridademae = pd.DataFrame(columns=['Data', 'Peso_medio_masc', 'Peso_medio_fem', 'Escolaridade_mae'])

for datanasc in datas:

    df_data = pd.DataFrame(sinasc_raw[sinasc_raw['DTNASC'] == datanasc])

    df_m = df_data[df_data['SEXO'] == 'Masculino']
    df_f = df_data[df_data['SEXO'] == 'Feminino']

    peso_medio_M = df_m['PESO'].mean()
    peso_medio_F = df_f['PESO'].mean()

    nova_linha = {
        'Data': datanasc,
        'Peso_medio_masc': peso_medio_M,
        'Peso_medio_fem': peso_medio_F,
        'Escolaridade_mae': df_data['ESCMAE']
        }
    
    data_pesomedio_sexo_escolaridademae = pd.concat([data_pesomedio_sexo_escolaridademae, pd.DataFrame([nova_linha])], ignore_index=True)

data_pesomedio_sexo_escolaridademae
```
