```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/SINASC/SINASC_RO_2019.csv')

sinasc = pd.DataFrame(data)
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sns.boxplot(
    data = sinasc['IDADEMAE'],
    y = sinasc['IDADEMAE'],
    width = .1
)
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(
    data = sinasc['PESO'],
    y = sinasc['PESO'],
    width = .1
)
```

####  3) Faça um histograma da idade da mãe

```python
# existem ferramentas para histogramas tanto no pd quanto no sns

hist = sns.histplot(
    data = sinasc['IDADEMAE'],
    binwidth = 1,
    kde = True
)
hist.set_xlabel('Idade da Mãe')
hist.set_ylabel('')
```

#### 4) Faça um histograma do peso do bebê

```python
hist = sns.histplot(
    data = sinasc['PESO'],
    binwidth = 100,
    kde = True,
)
hist.set_xlabel('Peso do Bebê (g)')
hist.set_ylabel('')
```
